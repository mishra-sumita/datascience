```python
import graphlab
```

```python
# Limit number of worker processes. This preserves system memory, which prevents hosted notebooks from crashing.
graphlab.set_runtime_config('GRAPHLAB_DEFAULT_NUM_PYLAMBDA_WORKERS', 4)
```

```python
fraudinstances = graphlab.SFrame('fraud.csv')
```

```python
fraudinstances.head()
```

```python
train_data,test_data = fraudinstances.random_split(.8, seed=0)
```

```python
fraud_model = graphlab.logistic_classifier.create(train_data,
                                                     target='Fraud Instance',
                                                     features=[ 'Damaged Item' , 'Item Not Avaiable' ,
                                                               'Item Not In Stock', 'Product Care Plan', 'Claim Amount' ,'Registered Online',
                                                               'Age Group','Marital Status' ,'Owns a Vehicle', 'Accomodation Type'],
                                                     validation_set=test_data)
```

```python
graphlab.canvas.set_target('ipynb')
fraud_model.evaluate(test_data, metric='roc_curve')
fraud_model.show(view='Evaluation')
```
